```python
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

```python
unique_items = purchase_data['Item ID'].nunique()
average_price = '${:,.2f}'.format(purchase_data['Price'].mean())
number_purchases = purchase_data['Item Name'].count()
revenue = '${:,.2f}'.format(purchase_data['Price'].sum())
pymoli = (unique_items, average_price, number_purchases, revenue)
#columns = ('Number of Unique Items', 'Average Price', 'Number of Purchases', 'Total Revenue')
pymoli_summary_df = pd.DataFrame([pymoli], columns = ['Number of Unique Items', 'Average Price', 'Number of Purchases', 'Total Revenue'])
pymoli_summary_df
```

## Gender Demographics

```python
total_count2 = purchase_data.groupby(['Gender']).nunique()
total_count2['test'] = total_count2['SN']/tot_player * 100
total_count2['test'] = total_count2['test'].round(decimals=2)
gender_demo = total_count2.loc[:, ['SN', 'test']].rename(columns={'SN':"Total Count", "test":"Percentage of Players"})
gender_demo.sort_values(by=['Total Count'], ascending=False)
```

## Purchasing Analysis (Gender)

```python
gender_group = purchase_data.groupby(["Gender"])
purchase_count = gender_group["Item ID"].count()
avg_purchase_price = gender_group["Price"].mean()
total_purchase_price = gender_group['Price'].sum()
avg_total_price_per_person = total_purchase_price/total_count2['SN']


purchasing_analysis_df = pd.DataFrame({"Purchase Count": purchase_count, 'Avg Purchase':avg_purchase_price, 
                                       "Total Purchase Value": total_purchase_price,'Avg Total Purchase per Person':avg_total_price_per_person})

purchasing_analysis_df["Avg Purchase"] = purchasing_analysis_df["Avg Purchase"].map('${:,.2f}'.format)
purchasing_analysis_df["Total Purchase Value"] = purchasing_analysis_df["Total Purchase Value"].map('${:,.2f}'.format)
purchasing_analysis_df["Avg Total Purchase per Person"] = purchasing_analysis_df["Avg Total Purchase per Person"].map('${:,.2f}'.format)


purchasing_analysis_df
```

## Age Demographics

```python
# Create bins and group labels
bins = [0, 9.99, 13.99, 18.99, 23.99, 28.99, 33.99, 38.99, 1000]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Cut to make age ranges
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

#players_group = purchase_data.groupby(['SN'])

#purchase_data
#players_group.sort_values(by=['Item Name'], ascending=False)
#age_group = purchase_data.groupby('Age Range')
age_group_count = purchase_data['Age Range'].value_counts()
percentage_age_group = age_group_count/tot_player * 100

# Make a dataframe
age_demo = pd.DataFrame({"Total Count": age_group_count, "Percentage of Players": percentage_age_group})

# Sort and format dataframe
age_demo.sort_index().round(2)
```

## Purchasing Analysis (Age)

```python
# Cut the data by age
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

# Make calculations 
avg_purchase_price_age = purchase_data.groupby(["Age Range"]).mean()['Price']
total_purchase_price_age = purchase_data.groupby(["Age Range"]).sum()["Price"]
count_age = purchase_data.groupby(["Age Range"]).count()["Price"]
avg_total_purchase_per_person = total_purchase_price_age / age_demo["Total Count"]

# Make a dataframe
purchasing_analysis = pd.DataFrame({"Purchase Counts": count_age, "Average Purchase Price":avg_purchase_price_age,
                                   "Total Purchase Value":total_purchase_price_age, 
                                    "Avg Total Purchase per Person":avg_total_purchase_per_person})

purchasing_analysis['Purchase Counts'] = purchasing_analysis['Purchase Counts'].map("{:,}".format)
purchasing_analysis['Average Purchase Price'] = purchasing_analysis['Average Purchase Price'].map("${:,.2f}".format)
purchasing_analysis['Total Purchase Value'] = purchasing_analysis['Total Purchase Value'].map("${:,.2f}".format)
purchasing_analysis['Avg Total Purchase per Person'] = purchasing_analysis['Avg Total Purchase per Person'].map("${:,.2f}".format)
purchasing_analysis
```

## Top Spenders

```python
players_group = purchase_data.groupby(['SN'])
purchase_count_player = players_group["Item ID"].count()
avg_purchase_price_player = players_group['Price'].mean()
total_purchase_value_player = players_group['Price'].sum()
top_spend_df = pd.DataFrame({'Purchase Count':purchase_count_player,"Average Purchase Price":avg_purchase_price_player, 
                             'Total Purchase Value':total_purchase_value_player})
# top_spend_df

top_spend_df_sorted = top_spend_df.sort_values(by=['Total Purchase Value'], ascending=False)
top_spend_df_sorted['Average Purchase Price'] = top_spend_df_sorted['Average Purchase Price'].map('${:,.2f}'.format)
top_spend_df_sorted['Total Purchase Value'] = top_spend_df_sorted['Total Purchase Value'].map('${:,.2f}'.format)
top_spend_df_sorted.head()

# user_sorted.loc[:,["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
```

## Most Popular Items

```python
item_group = purchase_data.groupby(["Item ID", "Item Name"])
item_count = item_group['Item ID'].count()
avg_item = item_group['Price'].mean()
total_price_item = item_group['Price'].sum()
pop_items = pd.DataFrame({'Purchase Count':item_count, "Item Price":avg_item, "Total Purchase Value":total_price_item})
pop_items_sorted = pop_items.sort_values(by="Purchase Count",ascending = False)
pop_items_sorted['Item Price'] = pop_items_sorted['Item Price'].map('${:,.2f}'.format)
pop_items_sorted['Total Purchase Value'] = pop_items_sorted['Total Purchase Value'].map('${:,.2f}'.format)
pop_items_sorted.head()
# ID_group = purchase_data.groupby(['Item ID']).count()
# ID_group
```

## Most Profitable Items

```python

pop_items_sorted_profit = pop_items.sort_values(by="Total Purchase Value",ascending = False)

pop_items_sorted_profit['Item Price'] = pop_items_sorted_profit['Item Price'].map('${:,.2f}'.format)
pop_items_sorted_profit['Total Purchase Value'] = pop_items_sorted_profit['Total Purchase Value'].map('${:,.2f}'.format)

pop_items_sorted_profit.head()
```
